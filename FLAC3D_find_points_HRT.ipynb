```python
InteractiveShell.ast_node_interactivity = "all"
df_TRT2['joint']=np.nan
df_TRT_UP = df_TRT2.loc[:23].reset_index(drop = True)
df_TRT_DW = df_TRT2.loc[24:(24+23)].reset_index(drop = True)
df_TRT_DX = df_TRT2.loc[(24+23+1):(24+23+1+23)].reset_index(drop = True)
df_TRT_SX = df_TRT2.loc[(24+23+1+23+1):(24+23+1+23+1+23)].reset_index(drop = True)

df_TRT_UP
df_TRT_DW
df_TRT_DX
df_TRT_SX
```
